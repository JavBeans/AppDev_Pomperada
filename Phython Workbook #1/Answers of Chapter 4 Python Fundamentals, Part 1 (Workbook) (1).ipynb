```python
# Exercise 1:

# Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

# Enter Hours: 45
# Enter Rate: 10
# Pay: 475.0

def compute(hour,rate):
  if hour >= 40:
    hour_excess = hour - 40
    pay_excess = hour_excess * 1.5
    pay_excess = pay_excess * rate
    hour = 40
    payment = hour * rate
    return pay_excess+payment

  else:
    return pay_excess*payment

hour = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
x = compute(hour,rate)
print(f"Total Pay: {x}")
```
