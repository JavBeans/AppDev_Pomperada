```python
#Exercise 2:

# Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

# Enter Hours: 45
# Enter Rate: 10
# Pay: 475.0

hour = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))


if hour >= 40:
  hour_excess = hour - 40
  pay_excess = hour_excess * 1.5
  pay_excess = pay_excess * rate
  hour = 40
  payment = hour * rate
  print(f"Total Pay: {pay_excess+payment}")

else:
  print(f" Total Pay: {pay_excess*payment}")
```

```python
#CHALLENGE

# Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

# Score           Grade
# [ 97, 100]      1.0
# [ 93,  97)        1.25
# [ 89,  93)        1.5
# [ 85,  89)        1.75
# [ 81,  85)        2.0
# [ 77,  81)        2.25
# [ 74,  77)        2.5
# [ 0,   74)        3.0

a = int(input("Enter Score: "))

if (a >= 0 and a < 74):
  print("Grade: 3.0")

elif (a >= 74 and a < 77):
  print("Grade: 2.5")

elif (a >= 77 and a < 81):
  print("Grade: 2.25")

elif (a >= 81 and a < 85):
  print("Grade: 2.0")

elif (a >= 85 and a < 89):
  print("Grade: 1.75")

elif (a >= 89 and a < 93):
  print("Grade: 1.5")

elif (a >= 93 and a < 97):
  print("Grade: 1.25")

elif (a >= 97 and a <= 100):
  print("Grade: 1.0")
```
